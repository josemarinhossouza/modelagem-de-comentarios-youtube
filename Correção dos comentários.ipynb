# Correção dos Comentários do YouTube

Este notebook realiza a **limpeza e correção de texto** em comentários do YouTube antes da modelagem de tópicos.

## Fluxo geral
1. Carrega os comentários de um arquivo CSV
2. Aplica correções de encoding (caracteres corrompidos)
3. Converte emojis em texto descritivo
4. Salva o resultado em um novo CSV

## 1. Imports

````python
#BIBLIOTECAS NECESSÁRIAS
import pandas as pd
import emoji
````

## 2. Carregamento dos dados

Lê o CSV com os comentários brutos exportados do YouTube.  
Ajuste o nome do arquivo conforme necessário.

````python
# 2. Carregar dados csv
 
comentarios_do_youtube = pd.read_csv("ARQUIVO COM COMENTÁRIOS.csv", encoding='utf-8-sig')
````

## 3. Funções de correção e pipeline

Três funções são definidas e combinadas em um pipeline:

- **`correcao_nativa_python`** — corrige caracteres corrompidos usando encode/decode nativo do Python (tenta `mac_roman` e depois `cp1252`)
- **`correcao_com_ftfy`** — usa a biblioteca `ftfy` para corrigir erros comuns de codificação de caracteres
- **`transformar_emoji_em_texto`** — substitui cada emoji pelo seu texto descritivo em português (ex: 😂 → `:rosto_com_lágrimas_de_alegria:`)
- **`pipeline_completo`** — encadeia as três correções acima em sequência

Ao final, fica a critério qual das funções utilizar, alguam específica ou correção completa.

````python
# 3. Correções (encode e decode: correção nativa do python. ftfy: correção de erro de codificação de caracteres. Emoji: susbstituição do emoji por texto correspondente)
def correcao_nativa_python(comentario):
    try:
        return comentario.encode('mac_roman').decode('utf-8')
    
    except:
        try:
            return comentario.encode('cp1252').decode('utf-8')
        
        except:
            return comentario

def corrrecao_com_ftfy (comentario):

        try:
           return ftfy.fix_text(comentario)

        except:
            return comentario
    
def transformar_emoji_em_texto(comentario):
    
    if pd.notna(comentario):
        return emoji.demojize(comentario, language='pt')

def pipeline_completo (comentarios):

    comentarios = correcao_nativa_python(comentarios)
   
    comentarios = corrrecao_com_ftfy(comentarios)
   
    comentarios = transformar_emoji_em_texto(comentarios)
   
    return comentarios

comentarios_do_youtube["comentário_sem_emoji"] = comentarios_do_youtube["COLUNA COM COMENTÁRIOS"].apply(transformar_emoji_em_texto)

````

## 4. Exportação dos resultados

Salva o DataFrame com os comentários corrigidos em um novo CSV, mantendo o encoding `utf-8-sig` para compatibilidade com Excel.

````python
# Criar csv do dataframe limpo e corrigido

salvar_dataframe = comentarios_do_youtube.to_csv('ARQUIVO COM COMENTÁRIOS CORRIGIDOS.csv', index='False', encoding='utf-8-sig')
````
